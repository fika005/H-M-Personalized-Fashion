- **perceived_colour_master_id (int64):** Colour grouping with 20 categories 
- **perceived_colour_master_name (object):** Perceived colour master name
- **department_no (int64):** Department number with 299 unique values
- **department_name (object):** Department name
- **index_code (int64):** A product category with 10 unique values
- **index_name (object):** Product index name 
- **index_group_no (int64):** A product category with 5 unique values
- **index_group_name (object):** Index group name
- **section_no (int64):** Product section number
- **section_name (object):** Product section name
- **garment_group_no (int64):** Garment group number
- **garment_group_name (object):** Garment group name
- **detail_desc (object):** Detailed description of the product

### Articles Types:
It would be informative to look at the distribution of index groups to see what categories the mojority of our articles fall into. 

```python
index_group_counts = df['index_group_name'].value_counts().reset_index()
index_group_counts.columns = ['Index Group Name', 'Number of Articles']

total_articles = len(df)
index_group_counts['Percentage'] = index_group_counts['Number of Articles'] / total_articles * 100

index_group_counts = index_group_counts.sort_values('Number of Articles', ascending=False)

plt.style.use('seaborn')
plt.figure(figsize=(12, 8))

ax = sns.barplot(x='Index Group Name', y='Number of Articles', data=index_group_counts, 
                 palette='viridis')

plt.title('Distribution of Articles Across Index Groups', fontsize=20, pad=20)
plt.xlabel('Index Group Name', fontsize=14, labelpad=10)
plt.ylabel('Number of Articles', fontsize=14, labelpad=10)

plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)

for i, v in enumerate(index_group_counts['Number of Articles']):
    ax.text(i, v, f'{v:,}\n({index_group_counts["Percentage"].iloc[i]:.1f}%)', 
            ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```


As it can be seen in the above chart, the main article categories are Ledieswear, Baby/Children, Menswear, sport, and divided. The majority of our articles fall into Ladieswear category. The second major category is Baby/children. This shows that there is more diversity among ladies and children articles compared to men's, sport's, and divided articles.
***

Now let us see what are the subcategories of the index group names:

```python
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")
sns.set_palette("deep")

index_name_counts = df['index_name'].value_counts().reset_index()
index_name_counts.columns = ['Index Name', 'Number of Articles']
index_name_counts['Percentage'] = index_name_counts['Number of Articles'] / len(df) * 100

ax = sns.barplot(x='Index Name', y='Number of Articles', data=index_name_counts, 
                 order=index_name_counts.sort_values('Number of Articles', ascending=False)['Index Name'])

plt.title("Distribution of Articles Across Index Names", fontsize=20, pad=20)
plt.xlabel('Index Name', fontsize=14, labelpad=10)
plt.ylabel('Number of Articles', fontsize=14, labelpad=10)

plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)

for i, p in enumerate(ax.patches):
    percentage = f"{index_name_counts['Percentage'].iloc[i]:.1f}%"
    ax.annotate(f"{p.get_height():,}\n({percentage})", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=9)

plt.tight_layout()
plt.show()
```

***
To have a better understanding of the Distribution of major categories and their subcategories, we can show them on a barchart:

```python
gr_df = df.groupby(['index_group_name', 'index_name'])["article_id"].nunique().reset_index()

sns.set(style='whitegrid', font_scale=1.5)  # Increased font scale
colors = sns.color_palette("husl", 8)

p = sns.catplot(x="index_group_name", y="article_id", hue="index_name", kind="bar", data=gr_df, height=12, aspect=2.5, palette=colors)  # Increased height

p.fig.suptitle('Distribution of Articles Across Index Groups and Names', fontsize=24, y=1.02)  # Increased fontsize
p.set_xlabels('Index Group Name', fontsize=18)  # Increased fontsize
p.set_ylabels('Number of Articles', fontsize=18)  # Increased fontsize

plt.xticks(rotation=45, ha='right')

ax = p.facet_axis(0, 0)
ax.tick_params(axis='both', which='major', labelsize=14)  # Increased labelsize

for i, c in enumerate(ax.containers):
    labels = [f'{(v.get_height() / 1000):.1f}K' if i == 0 else '' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge', fontsize=12, padding=2)  # Increased fontsize

# Adjust legend
plt.legend(title='Index Name', title_fontsize='16', fontsize='14', bbox_to_anchor=(1.05, 1), loc='upper left', ncol=2)  # Increased fontsize

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(right=0.85)  # Make room for the legend

# Add a bit more space at the top
plt.subplots_adjust(top=0.95)
```

Notice that Baby/Children is divided into 4 different subcategories namely, Baby Sizes 50-98, Children Accessories and Swimwear, Children sizes 134-170, and Children Sizes 92-140. Furthermore, Ladieswear is subcategorized as Ladies Accessories, Ladieswear, and Lingeries/Tights.
***

There are more sub_sub_categories of index_names that can be found in the following charts:

```python
values = df['product_group_name'].value_counts().tolist()
labels = df['product_group_name'].unique().tolist()

plt.figure(figsize=(20, 8))
sns.set_style("whitegrid")
sns.set_palette("deep")

prod_grp_name = pd.DataFrame({"Product Group Name": labels, "Number of Articles": values})
prod_grp_name = prod_grp_name.sort_values(["Number of Articles"], ascending=False)

ax = sns.barplot(x='Product Group Name', y='Number of Articles', data=prod_grp_name)

plt.title("Distribution of Articles Across Product Groups", fontsize=20, pad=20)
plt.xlabel("Product Group Name", fontsize=14, labelpad=10)
plt.ylabel("Number of Articles", fontsize=14, labelpad=10)

plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)

total = len(df)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width() / 2
    y = p.get_y() + p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
values = df['product_type_name'].value_counts().tolist()
labels = df['product_type_name'].unique().tolist()

plt.figure(figsize=(25, 10))
sns.set_style("whitegrid")
sns.set_palette("deep")

prod_grp_name = pd.DataFrame({"Product Type": labels, "Number of Articles": values})
prod_grp_name = prod_grp_name.sort_values(["Number of Articles"], ascending=False)

ax = sns.barplot(x='Product Type', y='Number of Articles', data=prod_grp_name)

plt.title("Distribution of Articles Across Product Types", fontsize=20, pad=20)
plt.xlabel("Product Type", fontsize=14, labelpad=10)
plt.ylabel("Number of Articles", fontsize=14, labelpad=10)

plt.xticks(rotation=90, ha='center', fontsize=10)
plt.yticks(fontsize=10)

total = len(df)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width() / 2
    y = p.get_y() + p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

There are 131 product types that can be found in the above chart. Among them, the majority of the articles are in the vest top group. Bra and underwear tights have the second and third diversity rank among the articles respectively.

### Articles Colors:
Now that we know what types of products we have, let us take a look at the color distribution among the articles:

```python
values = df['perceived_colour_value_name'].value_counts()
data = pd.DataFrame({'Perceived Color': values.index, 'Number of Articles': values.values})
data = data.sort_values('Number of Articles', ascending=False)

plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.set_palette("deep")

ax = sns.barplot(x='Perceived Color', y='Number of Articles', data=data)

plt.title("Distribution of Articles by Perceived Color", fontsize=20, pad=20)
plt.xlabel("Perceived Color", fontsize=14, labelpad=10)
plt.ylabel("Number of Articles", fontsize=14, labelpad=10)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)

total = df.shape[0]
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    ax.annotate(percentage, (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

There are 8 different percieved colors from which 2 are unkown or unidentified. The majority of articles are of dark color. These 8 main colors are divided into 20 and 50 sub-colors that can be seen in the following charts:

```python
plt.figure(figsize=(16, 6))
sns.set_style("whitegrid")
sns.set_palette("deep")

prod_grp_name = df['perceived_colour_master_name'].value_counts().reset_index()
prod_grp_name.columns = ["Perceived Colour Master Name", "Number of Articles"]
prod_grp_name = prod_grp_name.sort_values("Number of Articles", ascending=False)

ax = sns.barplot(x="Perceived Colour Master Name", y="Number of Articles", data=prod_grp_name)

plt.title("Distribution of Articles by Perceived Colour Master Name", fontsize=20, pad=20)
plt.xlabel("Perceived Colour Master Name", fontsize=14, labelpad=10)
plt.ylabel("Number of Articles", fontsize=14, labelpad=10)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)

total = df.shape[0]
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    ax.annotate(percentage, (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```
