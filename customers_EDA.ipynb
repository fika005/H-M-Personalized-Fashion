```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_customer = pd.read_csv('customers.csv')
df_customer['customer_id'] =\
    df_customer['customer_id'].apply(lambda x: int(x[-16:],16) ).astype('int64')


df_customer
```

```python
df_customer.info()
```

# FN and Active

```python
df_cust_backup = df_customer.copy()

df_customer[['FN','Active']] = df_customer[['FN','Active']].fillna(0)
```

```python
labels = ['No-FN', 'FN']
sizes = df_customer['FN'].value_counts()
sizes_percentage = sizes / sizes.sum() * 100

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(labels, sizes_percentage, color=['#3498db', '#e74c3c'])

ax.set_ylabel('Percentage', fontsize=12, fontweight='bold')
ax.set_xlabel('FN Status', fontsize=12, fontweight='bold')
ax.set_title('FN Status Distribution', fontsize=16, fontweight='bold')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

for i, bar in enumerate(bars):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.1f}%', ha='center', va='bottom', fontsize=10)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()
```

```python
labels = ['Not-Active', 'Active']
sizes = df_customer['Active'].value_counts()
sizes_percentage = sizes / sizes.sum() * 100

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(labels, sizes_percentage, color=['#3498db', '#e74c3c'])

ax.set_ylabel('Percentage', fontsize=12, fontweight='bold')
ax.set_xlabel('Active Status', fontsize=12, fontweight='bold')
ax.set_title('Active Status Distribution', fontsize=16, fontweight='bold')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

for i, bar in enumerate(bars):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.1f}%', ha='center', va='bottom', fontsize=10)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()
```

# Club Member Status

```python
df_customer['club_member_status'].unique()
```

```python
df_customer[['club_member_status']] = df_customer[['club_member_status']].fillna('N/A')
```

```python
plt.figure(figsize=(10, 6))
sizes = df_customer['club_member_status'].value_counts()
sizes_percentage = sizes / sizes.sum() * 100
status = df_customer['club_member_status'].unique()

bars = plt.bar(status, sizes_percentage, color=['#3498db', '#e74c3c', '#2ecc71', '#f39c12'])

plt.ylabel('Percentage', fontsize=12, fontweight='bold')
plt.xlabel('Member Status', fontsize=12, fontweight='bold')
plt.title('Club Member Status Distribution', fontsize=16, fontweight='bold')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.1f}%', ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()

print(sizes)
```

# Fashion News Frequency

```python
df_customer['fashion_news_frequency'].unique()
```

```python
# Cleaning up data NONE = None and nan = None
df_customer[['fashion_news_frequency']] = df_customer[['fashion_news_frequency']].fillna('None')
df_customer['fashion_news_frequency'] = df_customer['fashion_news_frequency'].replace('NONE', 'None')

df_customer['fashion_news_frequency'].unique()
```

```python
%matplotlib inline
plt.figure(figsize=(12, 6))
sizes = df_customer['fashion_news_frequency'].value_counts()
sizes_percentage = sizes / sizes.sum() * 100
frequency = df_customer['fashion_news_frequency'].unique()

bars = plt.bar(frequency, sizes_percentage, color=['#3498db', '#e74c3c', '#2ecc71'])

plt.ylabel('Percentage', fontsize=12, fontweight='bold')
plt.xlabel('Fashion News Frequency', fontsize=12, fontweight='bold')
plt.title('Distribution of Fashion News Frequency', fontsize=16, fontweight='bold')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.1f}%', ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=0, ha='center', fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()

print(sizes)
```
